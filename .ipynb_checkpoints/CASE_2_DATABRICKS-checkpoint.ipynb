```
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```
file_location = "/FileStore/tables/data_municipio-5.csv"
file_type = "csv"


schemaDDL = "TRANSACAO FLOAT, MUNICIPIO STRING, ESTADO STRING, DATA_ATUALIZACAO STRING"
infer_schema = "false"
first_row_is_header = "true"
delimiter = ";"


df = spark.read.format(file_type).schema(schemaDDL) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
w = Window().orderBy("MUNICIPIO")
df2 = df.withColumn("ORDEM_TRANSACAO", row_number().over(w))
```
